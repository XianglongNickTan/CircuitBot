```python
import numpy as np 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import os
import pickle
from pylab import mpl
import pylab
```

```python
file = open('../optimisation/utlis_data/exp_1_mean_10_trials', 'rb')
with_mean = pickle.load(file)
file.close()

file = open('../optimisation/utlis_data/exp_1_no_mean_10_trials', 'rb')
no_mean = pickle.load(file)
file.close()
```

```python
mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
plt.rcParams['figure.figsize'] = (6.0, 4.0)
plt.rcParams['image.interpolation'] = 'nearest' # 设置 interpolation style
plt.rcParams['image.cmap'] = 'winter' # 设置 颜色 style
plt.rcParams['savefig.dpi'] = 500 #图片像素
plt.rcParams['figure.dpi'] = 500 #分辨率

params = {
        "font.size": 12,     # 全局字号
        'font.family':'STIXGeneral', # 全局字体
        "figure.subplot.wspace":0.2, # 图-子图-宽度百分比
        "figure.subplot.hspace":0.4, # 图-子图-高度百分比
        "axes.spines.right":True,  # 坐标系-右侧线
        "axes.spines.top":True,   # 坐标系-上侧线
        "axes.titlesize":12,   # 坐标系-标题-字号
        "axes.labelsize": 12,  # 坐标系-标签-字号
        "legend.fontsize": 7,  # 图例-字号
        "xtick.labelsize": 10,  # 刻度-标签-字号
        "ytick.labelsize": 10,  # 刻度-标签-字号
        "xtick.direction":'out',   # 刻度-方向
        "ytick.direction":'out',  # 刻度-方向
        'xtick.major.size': 2,
        'ytick.major.size': 2,
        'axes.linewidth': 0.5,
        'xtick.major.pad': 3.5,
        'ytick.major.pad': 3.5,
        'axes.labelpad': 5.0,
        'axes.titlepad': 8.0,
        'axes.titlesize': 14.0,

}
pylab.rcParams.update(params)            #set figure parameter
pylab.rcParams.keys()
```

```python
prob_11 = []
prob_21 = []
prob_31 = []
prob_41 = []
prob_51 = []

for i in range(10):
    prob_11.append(no_mean.circle_prob[i][0])
    prob_21.append(no_mean.circle_prob[i][1])
    prob_31.append(no_mean.circle_prob[i][2])
    prob_41.append(no_mean.circle_prob[i][3])
    prob_51.append(no_mean.circle_prob[i][4])
```

```python
prob_12 = []
prob_22 = []
prob_32 = []
prob_42 = []
prob_52 = []

for i in range(10):
    prob_12.append(with_mean.circle_prob[i][0])
    prob_22.append(with_mean.circle_prob[i][1])
    prob_32.append(with_mean.circle_prob[i][2])
    prob_42.append(with_mean.circle_prob[i][3])
    prob_52.append(with_mean.circle_prob[i][4])
```

```python
fig,ax = plt.subplots()
ax.figsize=(48,48)

x_axis = np.linspace(1,10, 10)
x = range(1,11,1)

ins0=ax.plot(x_axis,prob_11, label = 'S1', ls='-.', marker='^', linewidth=0.8)
ins1=ax.plot(x_axis,prob_21, label = 'S2', ls='-.',marker='x' , linewidth=0.8)
ins2=ax.plot(x_axis,prob_31, label = 'S3', ls='-.',marker='.', linewidth=0.8)
ins3=ax.plot(x_axis,prob_41, label = 'S4', ls='-.',marker='.' , linewidth=0.8)
ins4=ax.plot(x_axis,prob_51, label = 'S5', ls='-.',marker='.' , linewidth=0.8)

ax.margins(0.1)
lns = ins0+ins1+ins2+ins3+ins4
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, ncol=1, frameon=False, loc="upper left")#loc="lower right" 图例右下角 , bbox_to_anchor=(1, 1.1), handlelength=2
# ax.set_xlabel("Experiemt Number", labelpad=50)
ax.set_xlabel("Experiment Number")
ax.set_ylabel("Probability")
ax.set_title("Circle Probability with Unnormalized Reward")
ax.set_xticks(x)

fig.savefig('Exp_1/Normal_nonormal/Nonormal circle probability.pdf', dpi=500,bbox_inches = 'tight')
fig.savefig('Exp_1/Normal_nonormal/Nonormal circle probability.svg', dpi=500,bbox_inches = 'tight')
fig.savefig('Exp_1/Normal_nonormal/Nonormal circle probability.svg', dpi=500,bbox_inches = 'tight')
```
