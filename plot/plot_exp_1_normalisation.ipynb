```python
fig,ax = plt.subplots()


c_0 = 11.3
c_1 = []
c_2 = []
c_3 = []
c_4 = []
c_5 = []

for i in range(40):
    if data.circle_num[i] == 1:
        c_1.append(data.voltage[i])
    elif data.circle_num[i] == 2:
        c_2.append(data.voltage[i])
    elif data.circle_num[i] == 3:
        c_3.append(data.voltage[i])
    elif data.circle_num[i] == 4:
        c_4.append(data.voltage[i])
    elif data.circle_num[i] == 5:
        c_5.append(data.voltage[i])
    
circle_mean = np.zeros(6)
circle_mean[0] = np.mean(c_0)
circle_mean[1] = np.mean(c_1)
circle_mean[2] = np.mean(c_2)
circle_mean[3] = np.mean(c_3)
circle_mean[4] = np.mean(c_4)
circle_mean[5] = np.mean(c_5)

circle_var = np.zeros(6)
circle_var[0] = np.var(c_0)
circle_var[1] = np.var(c_1)
circle_var[2] = np.var(c_2)
circle_var[3] = np.var(c_3)
circle_var[4] = np.var(c_4)
circle_var[5] = 0.2


x_axis = range(0,6,1)
       
ax = plt.subplot(111)


ax.errorbar(x_axis,circle_mean,yerr=circle_var,fmt='.',ecolor='b',color='b',elinewidth=1,
            capsize=2)
#fmt :   'o' ',' '.' 'x' '+' 'v' '^' '<' '>' 's' 'd' 'p'

ax.set_xlabel("Circle Number")
ax.set_ylabel("Voltage (V)")
# ax.set_title("Probability of Circles")

fig.savefig('Exp_1/Circle_probability/Circle number vs voltage.pdf', dpi=500,bbox_inches = 'tight')
fig.savefig('Exp_1/Circle_probability/Circle number vs voltage.svg', dpi=500,bbox_inches = 'tight')
fig.savefig('Exp_1/Circle_probability/Circle number vs voltage.jpg', dpi=500,bbox_inches = 'tight')
```
